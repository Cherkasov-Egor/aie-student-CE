```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    RocCurveDisplay,
)

import matplotlib.pyplot as plt
```

```python
data_path = "S05-hw-dataset.csv"
df = pd.read_csv(data_path)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["default"].value_counts(normalize=True)
```

В датасете около 3000 наблюдений и несколько десятков числовых признаков.
Целевая переменная `default` умеренно несбалансирована (~40% дефолтов).
Пропусков в данных не обнаружено, все признаки имеют числовой тип.


```python
X = df.drop(columns=["default", "client_id"])
y = df["default"]
```

`client_id` исключён из признаков, так как является техническим идентификатором
и не несёт предсказательной информации.


```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y,
)
```

```python
dummy = DummyClassifier(strategy="most_frequent", random_state=42)
dummy.fit(X_train, y_train)

y_pred_dummy = dummy.predict(X_test)
y_proba_dummy = dummy.predict_proba(X_test)[:, 1]

dummy_acc = accuracy_score(y_test, y_pred_dummy)
dummy_roc = roc_auc_score(y_test, y_proba_dummy)

dummy_acc, dummy_roc
```

```python
pipe = Pipeline(
    [
        ("scaler", StandardScaler()),
        ("logreg", LogisticRegression(max_iter=1000)),
    ]
)
```

```python
C_values = [0.01, 0.1, 1.0, 10.0]

results = []

for C in C_values:
    pipe.set_params(logreg__C=C)
    pipe.fit(X_train, y_train)

    y_pred = pipe.predict(X_test)
    y_proba = pipe.predict_proba(X_test)[:, 1]

    results.append(
        {
            "C": C,
            "accuracy": accuracy_score(y_test, y_pred),
            "roc_auc": roc_auc_score(y_test, y_proba),
        }
    )

results_df = pd.DataFrame(results)
results_df
```

```python
best_C = results_df.sort_values("roc_auc", ascending=False).iloc[0]["C"]

pipe.set_params(logreg__C=best_C)
pipe.fit(X_train, y_train)

RocCurveDisplay.from_estimator(pipe, X_test, y_test)
plt.title(f"ROC curve (LogReg, C={best_C})")
plt.show()
```

```python
comparison = pd.DataFrame(
    {
        "model": ["DummyClassifier", "LogisticRegression"],
        "accuracy": [dummy_acc, results_df.loc[results_df["C"] == best_C, "accuracy"].values[0]],
        "roc_auc": [dummy_roc, results_df.loc[results_df["C"] == best_C, "roc_auc"].values[0]],
    }
)

comparison
```

В рамках эксперимента была построена бейзлайн-модель и логистическая регрессия.
Бейзлайн показал ROC-AUC 0.5, что ожидаемо для модели, не использующей
информацию из признаков.

Логистическая регрессия существенно превзошла бейзлайн по ROC-AUC и accuracy,
что говорит о наличии информативного сигнала в данных. Подбор коэффициента
регуляризации показал, что слишком сильная регуляризация ухудшает качество,
а при умеренных значениях C метрики стабилизируются.

Логистическая регрессия в виде пайплайна со стандартизацией признаков является
разумным базовым решением для данной задачи бинарной классификации.
